```python

# 2️⃣ Load dataset
file_path = "/content/Airline_dataset.csv"  # Uploaded in Colab
df = pd.read_csv(file_path)

# 3️⃣ Data Preparation
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
df['Year'] = df['FL_DATE'].dt.year
df['Month'] = df['FL_DATE'].dt.month
df['route'] = df['ORIGIN_AIRPORT'] + "-" + df['DEST_AIRPORT']

# 4️⃣ Demand proxy: Monthly flights per route
agg = df.groupby(['route', pd.Grouper(key='FL_DATE', freq='MS')])['FL_DATE'].count().reset_index(name='flights')

# 5️⃣ Top 5 busiest routes
top_routes = agg.groupby('route')['flights'].sum().sort_values(ascending=False).head(5).index.tolist()
agg_top = agg[agg['route'].isin(top_routes)]

plt.figure(figsize=(10,5))
sns.lineplot(data=agg_top, x='FL_DATE', y='flights', hue='route')
plt.title("Monthly Flight Counts - Top Routes")
plt.show()

# 6️⃣ Delay Performance Analysis
delay_data = df[df['route'].isin(top_routes)].groupby(['route', pd.Grouper(key='FL_DATE', freq='MS')])['ARR_DELAY'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=delay_data, x='FL_DATE', y='ARR_DELAY', hue='route')
plt.axhline(0, color='red', linestyle='--')
plt.title("Average Arrival Delay (minutes) - Top Routes")
plt.show()

# 7️⃣ Forecasting
from prophet import Prophet
forecast_results = {}

for route in top_routes:
    ts = agg[agg['route'] == route][['FL_DATE', 'flights']].rename(columns={'FL_DATE': 'ds', 'flights': 'y'})
    model = Prophet(yearly_seasonality=True)
    model.fit(ts)

    future = model.make_future_dataframe(periods=12, freq='MS')
    forecast = model.predict(future)

    forecast_results[route] = forecast

    fig = model.plot(forecast)
    plt.title(f"Forecast - {route}")
    plt.xlabel("Date")
    plt.ylabel("Flights (Demand Proxy)")
    plt.show()

    forecast.to_csv(f"/content/{route.replace('/', '_')}_forecast.csv", index=False)

# 8️⃣ Prepare CSV for Tableau
# - One table for demand trends
agg.to_csv("/content/airline_monthly_flights.csv", index=False)

# - One table for delays
df.groupby(['route', pd.Grouper(key='FL_DATE', freq='MS')]) \
  .agg(avg_arr_delay=('ARR_DELAY', 'mean'),
       avg_dep_delay=('DEP_DELAY', 'mean'),
       flights=('FL_DATE', 'count')) \
  .reset_index() \
  .to_csv("/content/airline_delay_summary.csv", index=False)

print("✅ All analysis done. Forecasts, trend CSVs, and delay summaries saved for Tableau.")
```
